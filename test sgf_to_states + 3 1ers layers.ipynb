```python
import numpy as np
```

```python
def get_board(state,taille):
    planes = np.zeros((3, taille,taille))
    x=0
    while x<taille:
        y=0
        while y<taille:
            case=repr(state[x+1,y+1]) # conversion en string sinon le test == ne marche pas
            if case=='Black': # pion noir
                planes[0, x, y] = 1 
            if case=='White': #pion blanc
                planes[1, x, y] = 1 
            if case=='Empty': #vide
                planes[2, x, y] = 1 
            y+=1
        x+=1
    
    return planes
```
